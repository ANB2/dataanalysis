```python
samsungData = pd.read_csv('../data/samsungData.csv')
samsungData = samsungData.drop(['Unnamed: 0'], axis=1)
```

```python
subj1 = samsungData[samsungData['subject'] == 1]

numericActivity = pd.Categorical.from_array(subj1['activity'])

f, (ax1, ax2) = subplots(ncols=2)
f.set_size_inches(10, 5)

for lb, cl in zip(list(numericActivity.levels), 'b g r c m y k'.split()):
    idx = subj1['activity'] == lb
    ax1.scatter(subj1.index[idx], subj1.ix[idx,0], c=cl, label=lb)
    ax2.scatter(subj1.index[idx], subj1.ix[idx,1], c=cl, label=lb)

ax1.set_ylabel(samsungData.columns[0])
ax2.set_ylabel(samsungData.columns[1])
ax2.legend(loc='lower right')

f.tight_layout();
```

```python
numericActivity.levels
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.cluster.hierarchy import linkage, dendrogram

distanceMatrix = pdist(subj1.ix[:,:3])
dendrogram(linkage(distanceMatrix, method='complete'), 
           color_threshold=0.3, 
           leaf_label_func=lambda x: 'O' * (numericActivity.labels[x] + 1),
           leaf_font_size=6)

f = gcf()
f.set_size_inches(8, 4);
```

```python
f, (ax1, ax2) = subplots(ncols=2)
f.set_size_inches(10, 5)

for lb, cl in zip(list(numericActivity.levels), 'b g r c m y k'.split()):
    idx = subj1['activity'] == lb
    ax1.scatter(subj1.index[idx], subj1.ix[idx,9], c=cl, label=lb)
    ax2.scatter(subj1.index[idx], subj1.ix[idx,10], c=cl, label=lb)

ax1.set_ylabel(samsungData.columns[0])
ax2.set_ylabel(samsungData.columns[1])
ax2.legend(loc='lower right')

f.tight_layout();
```

```python
distanceMatrix = pdist(subj1.ix[:,9:12])
dendrogram(linkage(distanceMatrix, method='complete'), 
           color_threshold=0.3, 
           leaf_label_func=lambda x: 'O' * (numericActivity.labels[x] + 1),
           leaf_font_size=6)

f = gcf()
f.set_size_inches(8, 4);
```

```python
# a simple scale function to normalize a matrix
def scale(matrix):
    from numpy import mean, std
    return (matrix - mean(matrix, axis=0)) / std(matrix, axis=0)
```

```python
U, D, Vt = np.linalg.svd(subj1.ix[:,:-2].apply(scale), full_matrices=False)

f, (ax1, ax2) = subplots(ncols=2)
f.set_size_inches(10, 5)

for lb, cl in zip(list(numericActivity.levels), 'b g r c m y k'.split()):
    idx = subj1['activity'] == lb
    ax1.scatter(subj1.index[idx], U[idx,0], c=cl, label=lb)
    ax2.scatter(subj1.index[idx], U[idx,1], c=cl, label=lb)
    
ax1.set_ylabel('U[:,0]')
ax2.set_ylabel('U[:,1]')
ax2.legend(loc='lower right')

f.tight_layout();
```

```python
plot(Vt[1,:], 'ok');
```

```python
maxContrib = np.where(Vt[1,:] == np.max(Vt[1,:]))

distanceMatrix = pdist(subj1.take(range(9,12) + [maxContrib[0][0]], axis=1))
dendrogram(linkage(distanceMatrix, method='complete'), 
           color_threshold=0.3, 
           leaf_label_func=lambda x: 'O' * (numericActivity.labels[x] + 1),
           leaf_font_size=6)

f = gcf()
f.set_size_inches(8, 4);
```

```python
samsungData.columns[maxContrib[0][0]]
```

```python
from scipy.cluster.vq import kmeans, vq

data = np.matrix(subj1.ix[:,:-2])

centers, _ = kmeans(data, 6, iter=100)
cluster, _ = vq(data, centers)

df = pd.crosstab(cluster, subj1['activity'])
df
```

```python
idmax = df.index[df['laying'] == df['laying'].max()]

plot(centers[idmax[0],:10], 'ok')
ylabel('Cluster Center');
```

```python
idmax = df.index[df['walk'] == df['walk'].max()]

plot(centers[idmax[0],:10], 'ok')
ylabel('Cluster Center');
```
