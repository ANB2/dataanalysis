```python
test_subjects = [27, 28, 29, 30] # what is required: [27, 28, 29, 30]
train_subjects = filter(lambda x: x not in test_subjects, np.unique(samsungData['subject'])) # what is required: [1, 3, 5, 6]

# take training and test set from requested subjects
#train = sitstand[sitstand['subject'].isin(train_subjects)]
#test = sitstand[sitstand['subject'].isin(test_subjects)]

train = samsungData[samsungData['subject'].isin(train_subjects)]
test = samsungData[samsungData['subject'].isin(test_subjects)]

#pca = PCA(n_components=3)
#pca.fit(train[features])

# training data
X = train[features] #pca.transform(train[features]) #isomap.transform(train[features])
y = le.transform(train[labels])

# test data
Xt = test[features] #pca.transform(test[features]) #isomap.transform(test[features])
yt = le.transform(test[labels])
```

### SVM -- single and bootstrap-combined classifiers

```python
from sklearn.svm import SVC
from sklearn.grid_search import GridSearchCV

tuned_parameters = [{'gamma': [1., .1, 1e-2], 'C': [1e-2, .1]}]

svm = GridSearchCV( SVC(), tuned_parameters, cv=3, verbose=1 ).fit(X, y)
```

```python
print 'Best parameters set found on development set:'
print svm.best_estimator_
```

```python
pred = svm.predict(Xt)
print 'prediction accuracy: %.4f' % (1 - (1. / len(yt) * sum( pred != yt )))
```

```python
repeat = 7
classifiers = []

# train SVM classifiers
for i in xrange(repeat):
    # generate bootstrap data
    if i == 0:
        tix = train.index #np.random.choice(train.index, size=len(train), replace=False)
    else:
        tix = np.random.choice(train.index, size=len(train), replace=True)
    this_train = train.ix[tix,:]
    
    # training data
    X = this_train[features] #isomap.transform(this_train[features])
    y = le.transform(this_train[labels])
    
    # set the parameters by cross-validation
    tuned_parameters = [{'gamma': [1., .1, 1e-2], 'C': [1e-2, .1, 10, 1000]}]
    clf = GridSearchCV( SVC(class_weight='auto'), tuned_parameters, cv=3, verbose=1 ).fit(X, y)
    
    classifiers.append(clf)
```

```python
[x.best_estimator_ for x in classifiers]
```

```python
preds = np.zeros((len(classifiers), len(yt)), dtype=int)

i = 0
for clf in classifiers:
    preds[i] = clf.predict(Xt)
    i += 1
```

```python
total_pred = [np.bincount(x).argmax() for x in preds.T]
print 'prediction accuracy: %.4f' % (1 - (1. / len(yt) * sum( total_pred != yt )))
```
