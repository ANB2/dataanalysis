```python
import pandas as pd

dataset = pd.read_csv('https://spark-public.s3.amazonaws.com/dataanalysis/quiz3question4.csv ')
dataset.head()
```

```python
scatter(dataset['x'], dataset['y']);
```

```python
from scipy.cluster.vq import kmeans, vq

data = np.array(zip(dataset['x'], dataset['y']))

centers, _ = kmeans(data, 2)
cluster, _ = vq(data, centers)

scatter(dataset['x'], dataset['y'], c=cluster, cmap='autumn');
```
